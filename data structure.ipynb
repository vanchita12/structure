8.What is a hash table, and how does it relate to dictionaries in Python ?

   ->A hash table is a data structure that stores data in key-value pairs, where each key is unique and the hash table uses a hash function to compute an index into an array of buckets or slots, from which the desired value can be found.

10.Explain why strings are immutable in Python

 ->Strings are immutable in Python primarily for efficiency, security, and thread safety. Immutability ensures that a string's hash value remains constant, making them suitable as dictionary keys. It also prevents unintended modifications, enhances security, and allows for memory optimization and thread-safe operations.

12.Describe a scenario where using a tuple would be preferable over a list ?

  ->In scenarios where you need to represent data with fixed fields that should not be changed after creation, and where immutability enhances data integrity and semantic clarity, tuples are preferred over lists. This is particularly relevant for representing records, configurations, or any data where the structure and values should remain constant.

   -> Tuples are immutable in Python, meaning their elements cannot be modified after creation. This immutability provides data integrity, enables hashing for dictionary keys, and allows for memory optimization. If you need to modify data, use a list or create a new tuple with the desired changes

17.Describe the time complexity of accessing elements in a dictionary ?

  ->Accessing elements in a dictionary using a key has an average time complexity of O(1), also known as constant time. This means the access time is independent of the dictionary's size, making it very efficient for data retrieval. This is due to the use of hash tables for storing and retrieving data.

18.In what situations are lists preferred over dictionaries ?

Lists are preferred over dictionaries when:

Preserving the order of elements is important
Accessing elements by position is needed
Storing duplicate values is required
Iterating in a specific order is necessary
Working with homogeneous data is simpler

19.Why are dictionaries considered unordered, and how does that affect data retrieval

  -> Dictionaries are considered unordered because they do not guarantee the order of elements. This is due to their implementation using hash tables. Data retrieval is primarily based on keys, and the order of elements is not relevant for accessing values.

20.Explain the difference between a list and a dictionary in terms of data retrieval.

  ->Lists and dictionaries offer different approaches to data retrieval. Lists use index-based retrieval, while dictionaries use key-based retrieval. Dictionaries generally provide faster data retrieval for large datasets due to their use of hash tables. Choosing the appropriate data structure depends on the specific requirements of your task and how you need to access the data.
